## ðŸ”¹ What Are Python Dictionaries?

A **dictionary** in Python is a built-in data structure used to store data in the form of **keyâ€“value pairs**.

Each **key** is associated with a **value**, which allows fast and meaningful data access.

- Keys must be unique
- Keys are usually strings or numbers
- Values can be of any data type
- Dictionaries are mutable (can be changed after creation)

## ðŸ”¹ When to Use a Dictionary?

Use a dictionary when:
- You want **meaningful access** to data
- Each value is associated with a **label**
- Data represents an **object or record**
- Performance matters

## ðŸ”¹ When NOT to Use a Dictionary?

Avoid dictionaries when:
- You only need **ordered data**
- You need **index-based access**
- Keys are not meaningful

In such cases, use:
- **List** â†’ ordered collections
- **Tuple** â†’ fixed collections
- **Set** â†’ unique values


## ðŸ”¹ Dictionary Methods: `keys()` and `values()`

- **`dict.keys()`** â†’ Returns an iterable view containing all the keys in the dictionary.  
- **`dict.values()`** â†’ Returns an iterable view containing all the values in the dictionary.


````python
# calling the values and key through builtin function of dictionary
employes={1:"Ahmad",2:"ali",3:"malik",4:"hamid"}
print(employes.keys())
print(employes.values())
````

## ðŸ”¹ Dictionary Method: `copy()`

- **`dict.copy()`** â†’ Creates and returns a shallow copy of the dictionary without modifying the original dictionary.


````python
# copying the origional employes dictionary
employes={1:"Ahmad",2:"ali",3:"malik",4:"hamid"}
new_employes=employes.copy()
print(new_employes)
print(list(new_employes.keys()))
print(list(new_employes.values()))
````

## ðŸ”¹ Dictionary Method: `get()`

- **`dict.get(key, default=None)`** â†’ Returns the value of the specified key if it exists; otherwise returns the default value without raising an error.


````python
# geeting the necessary data form dictionary using get function
employes={1:"Ahmad",2:"ali",3:"malik",4:"hamid"}
print(new_employes.get(1)) # calling and printing the value at key 1
print(new_employes.get(2))
````

## ðŸ”¹ Dictionary Method: `pop()`

- **`dict.pop(key, default=None)`** â†’ Removes the specified key from the dictionary and returns its value; if the key does not exist, returns the default value (if provided) or raises a `KeyError`.


````python
# pop function in dictionary for removing the values
collouges={1:"john",2:"marry",3:"tom",4:"jerry"}
print(collouges)
removed=collouges.pop(1) # john remove actullay from the dict and stored in removed variable
print(removed)
print(collouges)
````

## ðŸ”¹ Dictionary Method: `popitem()`

- **`dict.popitem()`** â†’ Removes and returns the **last inserted key-value pair** as a tuple from the dictionary. Raises `KeyError` if the dictionary is empty.


````python
# popitem function remvoing the last item form the dictionary
removed_item=collouges.popitem() # this function actually remvoe the last item form the dict
print(collouges)
print(removed_item)
print(collouges)
````

## ðŸ”¹ Dictionary Method: `clear()`

- **`dict.clear()`** â†’ Removes all key-value pairs from the dictionary, leaving it empty.


````python
# clearing the whole dictinary  by using clear function
friends={1:"john",2:"marry",3:"tom",4:"jerry"}
print(friends)
removed=(friends.clear()) # it clear/erase the whole dictionary
print(removed)
````

````python
# looping the dictionary using the for loop
males={1:"john",2:"tom",3:"johni",4:"moony"}
for key,values in males.items():
  print(key,values)
````

## ðŸ”¹ Adding Items Using `update()`

- **`dict.update(other_dict)`** â†’ Adds key-value pairs from another dictionary.
- If the key exists, it **updates** the value; if the key does not exist, it **adds** a new item.


````python
# Existing dictionary
employee = {"name": "Khalid", "salary": 50000}

# Adding a single new item using update
employee.update({"department": "IT"})
print(employee)
# Output: {'name': 'Khalid', 'salary': 50000, 'department': 'IT'}

# Adding multiple items at once
employee.update({"location": "Lahore", "experience": 5})
print(employee)
# Output: {'name': 'Khalid', 'salary': 50000, 'department': 'IT', 'location': 'Lahore', 'experience': 5}
````


## ðŸ”¹ Dictionary vs List (Quick Comparison)

| Feature   | Dictionary   | List   |
|-----------|--------------|--------|
| Data access | By key | By index |
| Order | Insertion order (Python 3.7+) | Ordered |
| Lookup speed | Very fast (O(1)) | Slower |
| Best use case | Labeled data | Sequential data |


````python
# Get a view object of all keys in the 'males' dictionary
males={1:"john",2:"tom",3:"johni",4:"moony"}
keys_list = list(males.keys())  # Convert the keys view object into a list
print("Keys as a list:", keys_list) # Print the list of keys

# Get a view object of all values in the 'males' dictionary
values_list = list(males.values()) # Convert the values view object into a list
print("Values as a list:", values_list) # Print the list of values
````

````python
# Get a view object of all key-value pairs as tuples
items_list = list(males.items()) # Convert the items view object into a list of (key, value) tuples
print("Keys and values as a list of tuples:", items_list) # Print the list of key-value tuples
````

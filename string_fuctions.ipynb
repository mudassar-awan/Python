## 1Ô∏è‚É£ `.upper()` ‚Äì shout your text üì¢  
Converts **every** character to uppercase.

## 2Ô∏è‚É£ `.title()` ‚Äì title-case every word üé©  
First letter of **each word** becomes uppercase, the rest lowercase.

```python
message = 'hello'
print(message.title())
```

## 3Ô∏è‚É£ `.capitalize()` ‚Äì sentence case üìù  
Only the **very first** character is capitalised.

```python
message = 'hello'
print(message.capitalize())
```

## 4Ô∏è‚É£ `.swapcase()` ‚Äì flip the case üîÑ  
Upper ‚Üî lower, lower ‚Üî upper.

```python
message = 'hello'
print(message.swapcase())
```

## 5Ô∏è‚É£ `.lower()` ‚Äì whisper mode üîâ  
Everything in lowercase (safe for comparisons).

```python
message = "Hii"
print(message.lower())
```

## 6Ô∏è‚É£ `.split()` ‚Äì string ‚Üí list ‚úÇÔ∏è  
Splits on **whitespace** by default; returns a list of words.

```python
message = "Hello bro"
print(message.split())
```

## 7Ô∏è‚É£ `.strip()` ‚Äì vacuum the edges üßπ  
Removes leading & trailing **whitespace** (spaces, tabs, newlines).

```python
message = "\t hello\n "
print(f"'{message.strip()}'")
```

## 8Ô∏è‚É£ `.lstrip()` ‚Äì left-side trim ‚¨ÖÔ∏èüßπ  
Strips only the **left** edge.

```python
message = "\t Best of luck"
print(f"'{message.lstrip()}'")
```

## 9Ô∏è‚É£ `.rstrip()` ‚Äì right-side trim üßπ‚û°Ô∏è  
Strips only the **right** edge.

```python
message = "\t Best of luck   \n"
print(f"'{message.rstrip()}'")
```

## üîü `.find()` ‚Äì hunt for a substring üîç

```python
message = "what the hill"
print(message.find("w"))
```

## 1Ô∏è‚É£1Ô∏è‚É£ `.replace()` ‚Äì quick substitution üîÑ  
Replaces **all** occurrences of old ‚Üí new.

```python
message = "what the HILL"
print(message.replace('L', 'l'))
```

## 1Ô∏è‚É£2Ô∏è‚É£ `len()` ‚Äì how long is it? üìè  
Built-in function (not a method) that counts **characters**.

```python
message = 'message'
print(len(message))
```

## 1Ô∏è‚É£3Ô∏è‚É£ `.count()` ‚Äì letter census üî¢  
Counts how many times a substring appears.

```python
message = 'hello'
print(message.count('l'))
```

## 1Ô∏è‚É£4Ô∏è‚É£ `.startswith()` ‚Äì boolean gate üö™  
True if the string begins with the given substring.

```python
message = "Welocome to this notebook"
print(message.startswith('e'))
```

## 1Ô∏è‚É£5Ô∏è‚É£ `.endswith()` ‚Äì the grand finale ‚ùì  
True if the string ends with the given substring.

```python
message = 'do you practicing the python?'
print(message.endswith('?'))
```

## 1Ô∏è‚É£6Ô∏è‚É£ chaining methods ‚Äì swap & replace üîó  
Methods return **new** strings ‚Üí you can keep dot-joining them.

```python
message = "i practice python"
result = message.replace("python", "java")
print(result)
```

## 1Ô∏è‚É£7Ô∏è‚É£ `.join()` ‚Äì glue a list together üß¥  
Inserts the string **between** each item of an iterable.

```python
words = ['apple', 'banana', 'orange']
result = ",".join(words)
print(result)
```

## 1Ô∏è‚É£8Ô∏è‚É£ `.isalpha()` ‚Äì letters only üî§  
True if **all** characters are alphabetic (a-z, A-Z).

```python
num = "python"
print(num.isalpha())
```

## 1Ô∏è‚É£9Ô∏è‚É£ `.isdigit()` ‚Äì numbers only üî¢  
True if **all** characters are digits (0-9).

```python
num = '333'
print(num.isdigit())
```

## 2Ô∏è‚É£0Ô∏è‚É£ `.isalnum()` ‚Äì letters OR digits ‚úÖ  
True if **all** characters are alphanumeric (no spaces or punctuation).

```python
num = "hello123"
print(num.isalnum())
```

## 2Ô∏è‚É£1Ô∏è‚É£ `.isspace()` ‚Äì invisible characters üëª  
True if **every** character is whitespace.

```python
text = "      "
print(text.isspace())
```

## 2Ô∏è‚É£2Ô∏è‚É£ `.isupper()` ‚Äì all caps? üî†  
True if **all cased** characters are uppercase.

```python
message = "MESSAGE"
print(message.isupper())
```

## 2Ô∏è‚É£3Ô∏è‚É£ `.islower()` ‚Äì all lowercase? üî°  
True if **all cased** characters are lowercase.

```python
message = "python"
print(message.islower())
```

## 2Ô∏è‚É£4Ô∏è‚É£ `.removeprefix()` ‚Äì tidy front ‚úÇÔ∏è  
Removes the **exact** prefix if present (Python 3.9+).

```python
message = 'unhappy'
print(message.removeprefix("un"))
```

## 2Ô∏è‚É£5Ô∏è‚É£ `.removesuffix()` ‚Äì tidy back ‚úÇÔ∏è  
Removes the **exact** suffix if present (Python 3.9+).

```python
message = "happynot"
print(message.removesuffix('not'))
```

## 2Ô∏è‚É£6Ô∏è‚É£ slicing ‚Äì substring scalpel üî™  
Strings are **sequences** ‚Üí use `[start:stop:step]`.

```python
text = "python"
print(text[0:4])   # "pyth"
print(text[::-1])  # "nohtyp"
```

## 2Ô∏è‚É£7Ô∏è‚É£ quotes & escapes ‚Äì mix them safely üõ°Ô∏è  
Use backslash `\` or opposite quotes to avoid syntax errors.

```python
message = "Tim's cofee"
text = "hello\"malik\", how are you"
greetings = 'hello tim"s'
print(message)
print(text)
print(greetings)
```

## 2Ô∏è‚É£8Ô∏è‚É£ raw vs regular strings ‚Äì windows paths ü™ü  
Double backslash `\\` or raw string `r'...'` for file paths.

```python
hello = ('how\'s are you')
path = 'c\\mydirectory'
path1 = 'c/mydirectory'
print(hello)
print(path)
print(path1)
```

## 2Ô∏è‚É£9Ô∏è‚É£ indexing ‚Äì grab a single character üéØ  
Positive indices start at `0`, negative ones count from the end.

```python
message = " Hello World "
print(message[0])
print(message[1])
print(message[-1])
print(message[-2])
print(message[0:6])
print(message[6:])
print(message[0:])
print(message[7:])
```

## 3Ô∏è‚É£0Ô∏è‚É£ email dissection ‚Äì locate the goodies üìß  
Use `.find()` to get indices of username, domain, dot, etc.

```python
email = "user3232@gmail.com"
print("Index of '@':", email.find("@"))
print("Index of username:", email.find("user"))
print("Index of '.':", email.find("."))
print("Index of '.com':", email.find(".com"))
print("Index of domain:", email.find("gmail"))
print("Index of number:", email.find("3232"))
print("Slice from 3 ‚Üí end:", email[3:])
print("Slice 2-10:", email[2:10])
```

## 3Ô∏è‚É£1Ô∏è‚É£ visual index mapper ‚Äì character ‚Üî index üëÄ  
Loop over the exact span of any substring and print letter + index.

```python
email = "mudassar3232@gmail.com"
word = "mudassar"

start = email.find(word)
if start != -1:
    for i in range(start, start + len(word)):
        print(f"{email[i]} -> index {i}")
else:
    print("Word not found")
```

## 3Ô∏è‚É£2Ô∏è‚É£ list-comprehension style ‚Äì tuples galore üßÆ  
Same idea as above, but packed into a reusable list of `(char, index)` pairs.

```python
email = "mudassar3232@gmail.com"
word = "mudassar"

start = email.find(word)
if start != -1:
    letters_with_indices = [(email[i], i) for i in range(start, start + len(word))]
    print(letters_with_indices)
else:
    print("Word not found")
```

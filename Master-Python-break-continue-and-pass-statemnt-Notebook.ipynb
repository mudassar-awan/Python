## ğŸ”„ What Is a Loop?
- A **loop** lets you repeat a block of code while/for a condition is true.  
- **Two core flavours**:  
  - `while` â€“ repeats **until** a condition becomes `False`  
  - `for` â€“ repeats **over** an iterable (list, tuple, string, `range`, â€¦)  
- **Control keywords**:  
  - `break` â†’ exit the whole loop immediately  
  - `continue` â†’ skip the rest of *this* iteration and start the next  
  - `pass` â†’ do nothing (placeholder to avoid syntax errors)  

<a id="while-loop"></a>
## ğŸ”‚ `while` Loop â€“ â€œKeep sipping tea â˜• until the pot is empty!â€
```python
while condition_is_True:
    do_this()
```
âš ï¸ Make sure the condition eventually becomes `False` or youâ€™ll loop forever!

````python
# ---------- break statement demo with while loop  ----------
cup_of_tea = 0                 # counter starts at 0
while cup_of_tea <= 5:        # loop WHILE this is True
    print("I want more tea â˜•")
    cup_of_tea += 1            # increment counter by 1
    print(f"I have had {cup_of_tea} cups of tea")
    if cup_of_tea == 3:        # if we've had 3 cups of tea than exit loop
        print("Too much tea! Stopping now ğŸ›‘")
        break                  # exit the while loop immediately 
````

## ğŸ” `for` Loop & `range()` â€“ â€œServe 5 cups automaticallyâ€
```python
for variable in iterable:
    do_this()
```
`range(start, stop, step)` creates a sequence of numbers:  
- `start` inclusive (default 0)  
- `stop` **exclusive**  
- `step` default 1

````python
# ---------- for loop with range ----------
for cup in range(1, 6):       # 1,2,3,4,5  (6 is excluded)
    print(f"â˜• Chai is ready â†’ cup #{cup} served")
````

<a id="break"></a>
## ğŸ’” `break` â€“ â€œExit the loop right nowâ€
The break statement is used to stop a loop immediately, even if the loop condition is still true.
> stop printing once we hit 4

````python
# ---------- super-simple break ----------
for n in range(1, 10):
    if n == 4:
        print(f"ğŸ˜µ Encountered {n}, breaking out!")
        break                  # leave the loop
    print(n)
````

````python
for i in range(1, 6):
    if i == 3:
        break #-- exit the loop when i is 3
    print(i)
````

<a id="continue"></a>
## â­ï¸ `continue` â€“ â€œSkip this round, keep loopingâ€
The continue statement skips the current iteration of the loop and moves to the next iteration.
Print odd numbers only (skip evens)

````python
# ---------- super-simple continue ----------
for num in range(1, 8):
    if num % 2 == 0:          # even?
        continue              # skip  this iteration
    print(f"ğŸ² Odd number: {num}")
````

````python
for i in range(1, 6):
    if i == 3:
        continue # just skip this iteration when i is 3
    print(i)
````

<a id="pass"></a>
## â¸ï¸ `pass` â€“ â€œIâ€™ll fill this later, Python donâ€™t panic!â€
The pass statement does nothing. It is used as a placeholder when a statement is required syntactically but no action is needed.
> Useful for empty functions/loops/classes while prototyping.

````python
# ---------- super-simple pass ----------
for i in range(5):
    if i == 2:
        pass                   # nothing happens, just a placeholder
    else:
        print(f"i = {i}")

# empty function stub
def future_feature():
    pass   # TODO: implement later
````
